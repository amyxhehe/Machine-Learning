Hidden Layer 1

```python
x1 = 1
x2 = 3

#input 1 weights
w1 = 0.1
w2 = 0.2

#input 2 weights
w3 = 0.4
w4 = 0.3

#input 3 weights
w5 = 0.7
w6 = 0.6

h1_n1 = x1**w1 + x2**w2
h1_n2 = x1**w3 + x2**w4
h1_n3 = x1**w5 + x2**w6

print(h1_n1)
print(h1_n2)
print(h1_n3)
```

Hidden Layer 2

```python
x1 = h1_n1
x2 = h1_n2
x3 = h1_n3

#input 1 weights
w1 = 1.1
w2 = 1.2
w3 = 1.3

#input 2 weights
w4 = 1.4
w5 = 1.5
w6 = 1.6

#input 3 weights
w7 = 1.7
w8 = 1.8
w9 = 1.9

h2_n1 = x1 % w1 + x2 % w2 + x3 % w3
h2_n2 = x1 % w4 + x2 % w5 + x3 % w6
h2_n3 = x1 % w7 + x2 % w8 + x3 % w9

print(h2_n1)
print(h2_n2)
print(h2_n3)
```

Hidden Layer 3

```python
x1 = h2_n1
x2 = h2_n2
x3 = h2_n3

#input 1 weights
w1 = 1.01
w2 = 1.02
w3 = 1.03

#input 2 weights
w4 = 1.04
w5 = 1.05
w6 = 1.06

#input 3 weights
w7 = 1.07
w8 = 1.08
w9 = 1.09

h3_n1 = x1 % w1 + x2 % w2 + x3 % w3
h3_n2 = x1 % w4 + x2 % w5 + x3 % w6
h3_n3 = x1 % w7 + x2 % w8 + x3 % w9

print(h3_n1)
print(h3_n2)
print(h3_n3)
```

Hidden Layer 4

```python
x1 = h3_n1
x2 = h3_n2
x3 = h3_n3

#input 1 weights
w1 = 2.1
w2 = 2.2
w3 = 2.3

#input 2 weights
w4 = 2.4
w5 = 2.5
w6 = 2.6

h4_n1 = x1 % w1 + x2 % w2 + x3 % w3
h4_n2 = x1 % w4 + x2 % w5 + x3 % w6

print(h4_n1)
print(h4_n2)
```

Output Layer

```python
x1 = h4_n1
x2 = h4_n2

#weights
w1 = 0.25
w2 = 0.75

out_n = x1*w1 + x2*w2

print(out_n)
```
