Hidden Layer 1

```python
x1 = 5
x2 = 10
x3 = 15

#weights
w1 = 0.12
w2 = 0.24
w3 = 0.36

h1_n1 = x1*w1 - x2*w2 - x3*w3

print(h1_n1)
```

Hidden Layer 2

```python
x1 = h1_n1

#weights
w1 = 1.05
w2 = 1.06

h2_n1 = x1*w1
h2_n2 = x1*w2

print(h2_n1)
print(h2_n2)
```

Output Layer

```python
x1 = h2_n1
x2 = h2_n2

#weights
w1 = -1
w2 = -2

out_n = x1*w1 + x2*w2

print(out_n)
```
