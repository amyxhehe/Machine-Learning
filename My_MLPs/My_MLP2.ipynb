Hidden Layer 1

```python
x1 = 5

#weights
w1 = 0.3
w2 = 0.6

h1_n1 = x1 % w1
h1_n2 = x1 % w2

print(h1_n1)
print(h1_n2)
```

Hidden Layer 2

```python
x2 = h1_n1
x3 = h1_n2

#input 1 weights
w1 = 0.1

#input 2 weights
w2 = 0.2
w3 = 0.3

#input 3 weights
w4 = 0.4

h2_n1 = x2 % w1
h2_n2 = x1 % w2 + x3 % w3
h2_n3 = x3 % w4

print(h2_n1)
print(h2_n2)
print(h2_n3)
```

Hidden Layer 3

```python
x4 = h2_n1
x5 = h2_n2
x6 = h2_n3

#input 1 weights
w1 = 1.1
w2 = 1.2

#input 2 weights
w3 = 1.3

h3_n1 = x4 % w1 + x5 % w2
h3_n2 = x6 % w3

print(h3_n1)
print(h3_n2)
```

Output Layer

```python
x7 = h3_n1
x8 = h3_n2

#weights
w1 = 0.2
w2 = 0.4
w3 = 0.6

out_n = x7 % w1 + x5 % w2 + x8 % w3

print(out_n)
```
