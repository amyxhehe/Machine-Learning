Hidden Layer 1

```python
x1 = 2

w1 = 0.2

h1_n1 = x1*w1

print(h1_n1)
```

Hidden Layer 2

```python
x1 = h1_n1

#weights
w1 = 0.1
w2 = 0.3

h2_n1 = x1*w1
h2_n2 = x1*w2

print(h2_n1)
print(h2_n2)
```

Hidden Layer 3

```python
x1 = h2_n1
x2 = h2_n2

w1 = 0.5
w2 = 1.0

h3_n1 = x1*w1 + x2*w2

print(h3_n1)
```

Hidden Layer 4

```python
x1 = h3_n1

#weights
w1 = 0.15
w2 = 0.25
w3 = 0.35

h4_n1 = x1*w1
h4_n2 = x1*w2
h4_n3 = x1*w3

print(h4_n1)
print(h4_n1)
print(h4_n3)
```

Output Layer

```python
x1 = h4_n1
x2 = h4_n2
x3 = h4_n3

#weights
w1 = 0.14
w2 = 0.24
w3 = 0.34

out_n = x1*w1 + x2*w2 + x3*w3

print(out_n)
```
