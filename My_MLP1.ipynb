Hidden Layer 1

```python
x1 = 1
x2 = 2

#input 1 weights
w1 = 0.5
w2 = 1

#input 2 weights
w3 = 1.5
w4 = 2

#input 3 weights
w5 = 2.5
w6 = 3

h1_n1 = x1/w1 + x2/w2
h1_n2 = x1/w3 + x2/w4
h1_n3 = x1/w5 + x2/w6

print(h1_n1)
print(h1_n2)
print(h1_n3)
```

Hidden Layer 2

```python
x1 = h1_n1
x2 = h1_n2
x3 = h1_n3

#input 1 weights
w1 = 0.25
w2 = 1.25
w3 = 2.25

#input 2 weights
w4 = 0.75
w5 = 1.75
w6 = 2.75

#input 3 weights
w7 = 0.5
w8 = 1.5
w9 = 2.5

h2_n1 = x1/w1 + x2/w2 + x3/w3
h2_n2 = x1/w4 + x2/w5 + x3/w6
h2_n3 = x1/w7 + x2/w8 + x3/w9

print(h2_n1)
print(h2_n2)
print(h2_n3)
```

Hidden Layer 3

```python
x1 = h2_n1
x2 = h2_n2
x3 = h2_n3

#input 1 weights
w1 = 0.33
w2 = 1.33
w3 = 2.33

#input 2 weights
w4 = 0.66
w5 = 1.66
w6 = 2.66


h3_n1 = x1/w1 - x2/w2 - x3/w3
h3_n2 = x1/w4 - x2/w5 - x3/w6

print(h3_n1)
print(h3_n2)
```

Hidden Layer 4

```python
x1 = h3_n1
x2 = h3_n2

#input 1 weights
w1 = 0.25
w2 = 1.25
w3 = 2.25

#input 2 weights
w4 = 0.75
w5 = 1.75
w6 = 2.75

#input 3 weights
w7 = 0.1
w8 = 1
w9 = 2


h4_n1 = x1/w1 + x2/w2 + x3/w3
h4_n2 = x1/w4 + x2/w5 + x3/w6
h4_n3 = x1/w7 + x2/w8 + x3/w9

print(h4_n1)
print(h4_n2)
print(h4_n3)
```

Hidden Layer 5

```python
x1 = h4_n1
x2 = h4_n2
x3 = h4_n3

#input 1 weights
w1 = 0.25
w2 = 1.25
w3 = 2.25

#input 2 weights
w4 = 0.75
w5 = 1.75
w6 = 2.75

#input 3 weights
w7 = 0.5
w8 = 1.5
w9 = 2.5

h5_n1 = x1/w1 + x2/w2 + x3/w3
h5_n2 = x1/w4 + x2/w5 + x3/w6
h5_n3 = x1/w7 + x2/w8 + x3/w9

print(h5_n1)
print(h5_n2)
print(h5_n3)
```

Ouput Layer

```python
x1 = h5_n1
x2 = h5_n2
x3 = h5_n3

#input 1 weights
w1 = 0.33
w2 = 0.33
w3 = 0.33

#input 2 weights
w4 = 0.66
w5 = 0.66
w6 = 0.66


o_n1 = x1/w1 - x2/w2 - x3/w3
o_n2 = x1/w4 - x2/w5 - x3/w6

print(o_n1)
print(o_n2)
```
