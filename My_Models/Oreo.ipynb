```python
import math #to do math on single values
import numpy as np #to do math on arrays
import sympy as sp #to calculate derivatives
```

```python
class Oreo:
    #Implementing the Schwarzchild metric
    #In simpler words, calculating the spacetime interval of a Schwarzchild black hole
    #Schwarzchild black hole is a non charged and non rotating black hole

    def fit(self, X, y):
        self.X = X
        self.y = y
        #include all the mathematical calculations
        #constant
        G = 6.674 * 10**-11


        #Step 1: calculating the Mass of the Black Hole
        #calculating the mean of X
        #results in a single value each
        m1 = np.mean(X)
        m2 = np.mean(y)
        #calculating the deviation of X and y from m1 and m2 respectively
        #results in an array
        dev1 = X - m1
        dev2 = y - m2
        #calculating the element-wise product of both arrays
        #results in an array
        product_of_dev = np.multiply(dev1, dev2)
        #calculating the sum of the resulting array
        #results in a single value
        sum_of_product_of_dev = np.sum(product_of_dev)
        #calculating the square of deviation of X
        #results in an array
        squared_dev = np.square(dev1)
        #caculating the sum of the squared deviations
        #results in a single value
        sum_of_squared_dev = np.sum(squared_dev)
        #finally calculating the mass
        M = sum_of_product_of_dev / sum_of_squared_dev


        #Step 2: calculating the Azimuthal Angle of the Blach Hole
        #using the tan inverse formula
        azimuthal_angle = math.atan(X[1]/y[1])


        #Step 3: calculating the Speed of Light
        #we know its a constant but just for the sake of making a creative code
        #we decided to create our own speed of light
        #which actually turned out to be wayyy slower than the real one
        #but its alright
        r = 90.24
        #calculating velocity
        v = math.pi * r
        #caclulating energy
        E = 1/2 * M * v**2
        #calculating c from Einstein's mass-energy equation
        #E = M * (c**2)
        c = math.sqrt(E / M)


        #Step 4: calculating the Radial Distance from the center of the Black Hole
        x2 = 1.4
        y2 = 100
        #calculating the euclidean distance
        r = math.sqrt((x2-X[0])**2 + (y2-y[0])**2)


        #Step 5: calculating the Polar Angle
        #using trignometric functions
        angle = math.radians(180)
        z = r * math.cos(angle)
        polar_angle = math.atan((math.sqrt((X[1]**2) + (y[2]**2))) / z)


        #step 6: calculating the Time
        t = math.pi / math.sqrt(1-2*((G*M)/((c**2)*r)))


        #step 7: calculating the derivatives for our final equation
        wrt = sp.Symbol('x')
        #derivative of time
        t = int(t)*wrt
        dt = sp.diff(t, wrt)
        #derivative of polar angle
        polar_angle = int(polar_angle)*wrt
        dp = sp.diff(polar_angle, wrt)
        #derivative of azimuthal angle
        azimuthal_angle = int(azimuthal_angle)*wrt
        da = sp.diff(azimuthal_angle, wrt)
        #derivative of radial distance
        r = int(r)*wrt
        dr = sp.diff(r, wrt)


        #main formula
        spacetime_interval = -(1 - (2*G*M)/((c**2)*r)) * ((c**2) * (dt**2)) + ((1 - (2*G*M)/((c**2)*r))**-1) * dr**2 + (r**2)*(dp**2) + (r**2) * (math.sin(polar_angle)**2) * (da**2)

        #return the final value
        return spacetime_interval



    def predict(self, X, y):
        #send the data to be calculated
        prediction = self.fit(X,y)

        #return the output
        return prediction
```

```python
X = [1, 2, 3]
y = [4, 5, 6]

model = Oreo()

output1 = model.predict(X,y)

print(output1)
```
