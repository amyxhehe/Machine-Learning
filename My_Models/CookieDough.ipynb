```python
import math
import numpy as np
```

```python
class CookieDough:
    def __init__(self, b = 0.1):
        self.final_bias = b

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, d):
        data = d

        #Step 1
        new_data = []
        for x in data:
          w = x*0.2
          b = x*2
          new_data.append(x*w+b)

        #Step 2
        mean = sum(new_data)/len(new_data)

        #Step 3
        sig = self.sigmoid(mean)

        #Step 4
        new_num = int(sig*2)

        #Step 5
        fct = math.factorial(new_num)
        result = fct*self.final_bias*20

        return result


    def predict(self, d):
        result = self.fit(d)
        prediction = 0
        if result>3.5:
            prediction = 1
        else:
            prediction = 0

        return prediction
```

```python
X = [10,20,30,40,50]

model = CookieDough()

output = model.predict(X)

print(output)
```
